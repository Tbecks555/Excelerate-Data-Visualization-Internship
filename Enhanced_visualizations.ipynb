```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

# Data from the "Average CPR and CPC by Campaign" chart in your document
# Approximated values from the chart image and insight findings
data = {
    'Campaign Name': [
        'SHU_Students (Nigeria)',
        'SHU3_ (Students Apart from India and US)',
        'SHU_Students (India)',
        'SHU_Students (Nepal)',
        'SHU_Students (Ghana)',
        'SHU_6 (Educators and Principals)',
        'SHU_Students (UAE)',
        'SHU_Students (Canada)',
        'SHU_Students (USA)',
        'SHU_Students (UK)',
        'SHU_Students (Australia)'
    ],
    'Cost per Result (CPR)': [
        0.5, # Approximate value from the chart
        1.0, # Approximate value
        1.5, # Approximate value
        2.5, # Approximate value
        4.0, # Approximate value
        5.0, # Approximate value
        9.0, # Approximate value
        7.5, # Approximate value
        8.0, # Approximate value
        18.5, # Approximate value
        22.0 # Approximate value
    ],
    'Cost Per Click (CPC)': [
        0.2, # Approximate value from the chart
        0.4, # Approximate value
        0.6, # Approximate value
        1.0, # Approximate value
        2.5, # Approximate value
        2.0, # Approximate value
        4.5, # Approximate value
        5.5, # Approximate value
        5.0, # Approximate value
        6.0, # Approximate value
        8.0  # Approximate value
    ]
}

df = pd.DataFrame(data)

# Sort the DataFrame by 'Cost per Result (CPR)' in descending order for consistent sorting
df_sorted = df.sort_values(by='Cost per Result (CPR)', ascending=False)

# Define the four campaigns that are NOT performing well (Highest CPR & CPC Campaigns)
# Based on your "Insight findings"
not_performing_well_campaigns = [
    'SHU_Students (Australia)',
    'SHU_Students (UK)',
    'SHU_Students (UAE)',
    'SHU_Students (Canada)'
]

# Define colors
bright_not_performing_color_cpr = '#FF4500' # OrangeRed for not performing well CPR
bright_not_performing_color_cpc = '#FF8C00' # DarkOrange for not performing well CPC

# Dull/subtle colors for other (better performing) campaigns
other_campaigns_color_cpr = '#AEC6CF' # Muted blue for other CPR
other_campaigns_color_cpc = '#B0C4DE' # Light steel blue for other CPC


# Set a stylish font for the plot
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans']

# Create the grouped horizontal bar chart
fig, ax = plt.subplots(figsize=(14, 9)) # Adjust figure size for better readability
fig.set_facecolor('#F5F5F5') # Light grey background for the figure
ax.set_facecolor('#FFFFFF') # White background for the plot area

bar_width = 0.35 # Width of each individual bar

# Create an array for the y-axis positions of the bars
y_pos = np.arange(len(df_sorted['Campaign Name']))

# Plot 'Cost per Result (CPR)' bars
bars_cpr = []
for i, row in df_sorted.iterrows():
    campaign = row['Campaign Name']
    x_pos = y_pos[df_sorted.index.get_loc(i)] - bar_width/2
    if campaign in not_performing_well_campaigns:
        current_cpr_color = bright_not_performing_color_cpr
    else:
        current_cpr_color = other_campaigns_color_cpr
    bar = ax.barh(x_pos, row['Cost per Result (CPR)'], height=bar_width,
                  label='Cost per Result (CPR)' if i == df_sorted.index[0] else "",
                  color=current_cpr_color, edgecolor='white', linewidth=0.5)
    bars_cpr.append(bar)

# Plot 'Cost Per Click (CPC)' bars, offset slightly
bars_cpc = []
for i, row in df_sorted.iterrows():
    campaign = row['Campaign Name']
    x_pos = y_pos[df_sorted.index.get_loc(i)] + bar_width/2
    if campaign in not_performing_well_campaigns:
        current_cpc_color = bright_not_performing_color_cpc
    else:
        current_cpc_color = other_campaigns_color_cpc
    bar = ax.barh(x_pos, row['Cost Per Click (CPC)'], height=bar_width,
                  label='Cost Per Click (CPC)' if i == df_sorted.index[0] else "",
                  color=current_cpc_color, edgecolor='white', linewidth=0.5)
    bars_cpc.append(bar)


# Add title and labels with improved font sizes
ax.set_title('Campaign Performance: Cost Per Result (CPR) and Cost Per Click (CPC)', fontsize=18, pad=20, color='#333333')
ax.set_xlabel('Cost (INR)', fontsize=14, labelpad=15, color='#555555')
ax.set_ylabel('Campaign Name', fontsize=14, labelpad=15, color='#555555')

# Set x-axis limits to start from 0 and extend slightly beyond max value
max_val = max(df_sorted['Cost per Result (CPR)'].max(), df_sorted['Cost Per Click (CPC)'].max())
ax.set_xlim(0, max_val * 1.15) # Extend a bit for labels

# Set y-axis ticks and labels
ax.set_yticks(y_pos)
ax.set_yticklabels(df_sorted['Campaign Name'], fontsize=16, color='#555555')
plt.xticks(fontsize=16, color='#555555')

# Add subtle vertical gridlines for better readability of values
ax.grid(axis='x', linestyle='--', alpha=0.6, color='#CCCCCC')

# Remove top and right spines for a cleaner look
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#CCCCCC')
ax.spines['bottom'].set_color('#CCCCCC')

# Add direct value labels on the bars
# Labels for CPR bars
for i, bar_group in enumerate(bars_cpr):
    for bar in bar_group: # bar_group is a list of Rectangle objects
        ax.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2,
                f'{df_sorted["Cost per Result (CPR)"].iloc[i]:.1f}', va='center', ha='left',
                color='black', fontsize=9)

# Labels for CPC bars
for i, bar_group in enumerate(bars_cpc):
    for bar in bar_group:
        ax.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2,
                f'{df_sorted["Cost Per Click (CPC)"].iloc[i]:.1f}', va='center', ha='left',
                color='black', fontsize=9)


# Create a more informative and simplified legend
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor=bright_not_performing_color_cpr, label='Not Performing Well (CPR)'),
    Patch(facecolor=bright_not_performing_color_cpc, label='Not Performing Well (CPC)'),
    Patch(facecolor=other_campaigns_color_cpr, label='Other Campaigns (CPR)'),
    Patch(facecolor=other_campaigns_color_cpc, label='Other Campaigns (CPC)')
]
ax.legend(handles=legend_elements, loc='lower right', fontsize=10, frameon=True, edgecolor='#CCCCCC', facecolor='#FFFFFF')

# Invert y-axis to have the highest values at the top, matching the original chart's visual flow
ax.invert_yaxis()

# Improve layout to prevent labels from overlapping
plt.tight_layout()

# Display the plot
plt.show()
```
