```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
from matplotlib.patches import Patch

# Data setup
data = {
    'Campaign ID': [
        'Campaign 2',
        'Campaign 1',
        'Campaign 4',
        'Campaign 6',
        'Campaign 8',
        'Campaign 11',
        'Campaign 3',
        'Campaign 5',
        'Campaign 7',
        'Campaign 9',
        'Campaign 10'
    ],
    'CTR': [6.0, 2.8, 5.5, 4.2, 9.8, 8.8, 4.3, 3.5, 2.2, 2.5, 3.5],
    'Unique CTR': [6.2, 4.4, 5.9, 4.1, 9.3, 8.5, 4.2, 3.8, 4.0, 2.8, 3.4]
}

df = pd.DataFrame(data)
df_sorted = df.sort_values(by='CTR', ascending=True)

# Highlighted campaigns
not_performing_well = [
    'Campaign 9',
    'Campaign 10',
    'Campaign 3',
    'Campaign 1'
]

# Colors
color_ctr_highlight = '#FF4500'
color_uctr_highlight = '#FF8C00'
color_ctr_other = '#AEC6CF'
color_uctr_other = '#B0C4DE'

# Style
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans']

# Plot
fig, ax = plt.subplots(figsize=(14, 8))
fig.set_facecolor('#F5F5F5')
ax.set_facecolor('#FFFFFF')

bar_height = 0.35
ind = np.arange(len(df_sorted))

# CTR bars
bars_ctr = []
for i, row in df_sorted.iterrows():
    campaign = row['Campaign ID']
    y_pos = ind[df_sorted.index.get_loc(i)] - bar_height / 2
    color = color_ctr_highlight if campaign in not_performing_well else color_ctr_other
    bar = ax.barh(y_pos, row['CTR'], bar_height,
                  label='CTR in %' if i == df_sorted.index[0] else "",
                  color=color, edgecolor='white')
    bars_ctr.append(bar)

# Unique CTR bars
bars_uctr = []
for i, row in df_sorted.iterrows():
    campaign = row['Campaign ID']
    y_pos = ind[df_sorted.index.get_loc(i)] + bar_height / 2
    color = color_uctr_highlight if campaign in not_performing_well else color_uctr_other
    bar = ax.barh(y_pos, row['Unique CTR'], bar_height,
                  label='Unique CTR in %' if i == df_sorted.index[0] else "",
                  color=color, edgecolor='white')
    bars_uctr.append(bar)

# Labels & Titles
ax.set_title('Campaign Performance: CTR & Unique CTR', fontsize=22, pad=20, color='#333333')
ax.set_ylabel('Campaign Name', fontsize=20, labelpad=15, color='#555555')
ax.set_xlabel('CTR (%)', fontsize=20, labelpad=15, color='#555555')

ax.set_yticks(ind)
ax.set_yticklabels(df_sorted['Campaign ID'], fontsize=15, color='#555555')
plt.xticks(fontsize=13, color='#555555')

# Set limits
max_ctr = df_sorted[['CTR', 'Unique CTR']].max().max()
ax.set_xlim(0, max_ctr * 1.15)

# Grid and spines
ax.grid(axis='x', linestyle='--', alpha=0.6, color='#CCCCCC')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#CCCCCC')
ax.spines['bottom'].set_color('#CCCCCC')

# Value labels
def autolabel(bars, values):
    for i, bar_group in enumerate(bars):
        for bar in bar_group:
            width = bar.get_width()
            ax.text(width + 0.3, bar.get_y() + bar.get_height() / 2,
                    f'{values[i]:.1f}%', va='center', fontsize=9, color='#333333')

autolabel(bars_ctr, df_sorted['CTR'].tolist())
autolabel(bars_uctr, df_sorted['Unique CTR'].tolist())

# Legend
legend_elements = [
    Patch(facecolor=color_ctr_highlight, label='Not Performing Well (CTR)'),
    Patch(facecolor=color_uctr_highlight, label='Not Performing Well (Unique CTR)'),
    Patch(facecolor=color_ctr_other, label='Other Campaigns (CTR)'),
    Patch(facecolor=color_uctr_other, label='Other Campaigns (Unique CTR)')
]
ax.legend(handles=legend_elements, loc='lower right', fontsize=12, frameon=True, edgecolor='#CCCCCC', facecolor='#FFFFFF')

plt.tight_layout()
plt.show()
```

Underperforming campaigns based on CTR and Unique CTR are Campaign 3 (Australia), 10 (UK), 1 (Educators and Principals), and 9 (UAE), indicating weaker audience engagement compared to other campaigns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap, BoundaryNorm

# 1. Illustrative Data for CPC by Age & Geography:
#    This data is generated to reflect general trends from your overall CPC chart,
#    as specific CPC by Age & Geography data was not available in the provided images.
#    Lower CPC values indicate better efficiency.
data = {
    'Australia': [7.50, 9.00, 8.50, 6.00, 5.50, 5.00], # High CPC overall
    'Canada': [5.00, 6.50, 5.80, np.nan, np.nan, np.nan], # Moderate/High CPC overall
    'Ghana': [2.00, 3.00, 3.50, np.nan, np.nan, np.nan], # Low CPC overall
    'India': [0.50, 0.80, 1.00, np.nan, np.nan, np.nan], # Very Low CPC overall
    'Nepal': [0.80, 1.20, 1.50, np.nan, np.nan, np.nan], # Low CPC overall
    'Nigeria': [0.30, 0.40, 0.50, np.nan, np.nan, np.nan], # Very Low CPC overall
    'UAE': [4.00, 5.00, 4.80, np.nan, np.nan, np.nan], # Moderate CPC overall
    'UK': [5.50, 7.00, 6.50, np.nan, np.nan, np.nan], # High CPC overall
    'USA': [np.nan, 4.50, 5.20, np.nan, np.nan, np.nan] # Moderate CPC overall
}

# Original index (Age Groups) and columns (Geographies)
index_age_groups = ['13-17', '18-24', '25-34', '35-44', '45-54', '55-64']
columns_geography = list(data.keys())

df = pd.DataFrame(data, index=index_age_groups)

# Transpose the DataFrame to swap axes: Geographies become index, Age Groups become columns
df_transposed = df.T

# Calculate row and column averages for the transposed DataFrame.
df_with_averages_transposed = df_transposed.copy()
# Average for each Geography (now a row) across Age Groups (now columns)
df_with_averages_transposed['Average (Age Group)'] = df_with_averages_transposed.mean(axis=1)
# Average for each Age Group (now a column) across Geographies (now rows)
df_with_averages_transposed.loc['Average (Geography)'] = df_transposed.mean(axis=0)


# Find the minimum and maximum CPC values from the original data (before transposition and averages)
min_cpc = df.min().min()
max_cpc = df.max().max()

# 2. Define CPC Tiers and Colors for Categorical Coloring:
#    We define specific ranges for CPC and assign distinct colors.
#    Lower CPC is better (more efficient).

# Define boundaries for CPC ranges (inclusive lower, exclusive upper for most)
# The last bound ensures the maximum value is included.
bounds = [0, 1.0, 3.0, 5.0, 7.0, df_with_averages_transposed.max().max() + 0.1]

# Define colors for each range (from green for excellent to dark red for very high cost)
colors = ['#66BB6A',   # Excellent (Green)
          '#AED581',   # Good (Light Green)
          '#FFEB3B',   # Average (Yellow)
          '#FF9800',   # High (Orange)
          '#D32F2F']   # Very High (Dark Red)

# Create a custom colormap and normalization based on the defined bounds
cmap = ListedColormap(colors)
norm = BoundaryNorm(bounds, cmap.N)

# 3. Create the Heatmap Visualization:
plt.figure(figsize=(14, 9)) # Adjust figure size for better readability
ax = sns.heatmap(
    df_with_averages_transposed, # Use the transposed DataFrame
    annot=True,      # Display the CPC values in each cell
    fmt=".2f",       # Format the annotations to two decimal places
    cmap=cmap,       # Apply the custom categorical colormap
    norm=norm,       # Use the custom normalization for coloring
    linewidths=.5,   # Add thin white lines between cells for separation
    linecolor='white',
    cbar_kws={
        'label': 'Cost Per Click (CPC) in INR', # Label for the color bar
        'ticks': [b + (bounds[i+1]-b)/2 for i, b in enumerate(bounds[:-1])] # Position ticks in the middle of color bands
    },
    square=False     # Allow cells to be rectangular if needed by aspect ratio
)

# Customize color bar tick labels for clarity
cbar = ax.collections[0].colorbar
cbar.set_ticklabels(['Excellent (0-1)', 'Good (1-3)', 'Average (3-5)', 'High (5-7)', 'Very High (>7)'])


# 4. Enhance Annotations for Min/Max Values:
#    Iterate through the text annotations in the heatmap to highlight
#    the absolute minimum and maximum CPC values found in the original data.
for text in ax.texts:
    try:
        val = float(text.get_text())
        # Highlight the minimum CPC value in bold dark green
        if val == min_cpc:
            text.set_weight('bold')
            text.set_color('#006400') # Dark Green
        # Highlight the maximum CPC value in bold white (for contrast on dark red)
        elif val == max_cpc:
            text.set_weight('bold')
            text.set_color('white')
    except ValueError:
        # Handle cases where text might not be a float (e.g., NaN cells)
        pass

# Set plot title and axis labels
plt.title('Campaign Performance: Cost Per Click (CPC) by Geography & Age\n'
          ' (Lower CPC = More Efficient)', fontsize=18, pad=20)
plt.xlabel('Age Group', fontsize=14, labelpad=15) # X-axis is now Age Group
plt.ylabel('Geography', fontsize=14, labelpad=15) # Y-axis is now Geography

# Ensure X and Y axis labels are not rotated, as requested
plt.xticks(rotation=0, fontsize=11)
plt.yticks(rotation=0, fontsize=11)

# Adjust layout to prevent labels from overlapping and ensure everything fits
plt.tight_layout()
plt.show()
```

Poor cost efficiency is observed in older age groups in Australia and the UK, indicating higher costs with lower returns for these segments.
